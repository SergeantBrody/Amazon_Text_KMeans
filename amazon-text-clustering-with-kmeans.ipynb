```python
import numpy as np
import pandas as pd
data = pd.read_csv('/kaggle/input/amazon-ml-product-browse-node-classification/amazon dataset/amazon_train.csv')
df = pd.DataFrame(data)
```

```python
df
```

```python

df = df.drop('BROWSE_NODE_ID', axis=1)
```

```python
df = df.drop('BRAND', axis=1)
```

```python
df
```

```python
df.dropna(axis=0)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
```

```python
documents = df['DESCRIPTION'].values.astype("U")
```

```python
documents = df['BULLET_POINTS'].values.astype("U")
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
features = vectorizer.fit_transform(documents)
```

```python
k = 20
model = KMeans (n_clusters=k, init='k-means++', max_iter = 100, n_init=1)
model.fit(features)
```

```python
df['cluster'] = model.labels_
```

```python
df
```

```python
clusters = df.groupby('cluster')
for cluster in clusters.groups:
    f = open('cluster' + str(cluster)+ '.csv', 'w')
    data = clusters.get_group(cluster)[['TITLE', 'DESCRIPTION', 'BULLET_POINTS']]
    f.write(data.to_csv(index_label='id'))
    f.close()
```

```python
print ("Cluster centroids: \n")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names_out()

for i in range(k):
    print("Cluster %d: " % i)
    for j in order_centroids[i, :10]: 
        print (' %s' % terms[j])
    print('--------------')
```
